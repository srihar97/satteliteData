```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataset is in a CSV file named 'conjunction_data.csv'
file_path = '/content/sort-minRange.csv'

# Read the dataset
df = pd.read_csv(file_path, parse_dates=['TCA'])

# Add a new column for the day of the analysis
df['Analysis_Day'] = df['TCA'].dt.date

# Group data by day and calculate the total number of conjunctions each day
total_conjunctions_per_day = df.groupby('Analysis_Day').size()

# Plotting the total number of conjunctions over the analysis days
plt.figure(figsize=(10, 6))
total_conjunctions_per_day.plot(kind='line', marker='o', linestyle='-', color='b')
plt.title('Total Conjunctions Over Time')
plt.xlabel('Analysis Day')
plt.ylabel('Number of Conjunctions')
plt.grid(True)
plt.show()

```
